**Q9. What is the Difference Between a Primary Key and a Foreign Key?**

**Ans-**

*   **Primary Key**

    *   Uniquely identifies each row in a table.
    *   Cannot be NULL.
    *   Only one primary key per table.
    *   Example: student_id in a Students table.
*   **Foreign Key**

    *   Creates a link between two tables.
    *   References the primary key of another table.
    *   Can have multiple foreign keys in a table.
    *   Example: customer_id in Orders table referencing Customers table.

















**Q10. What Are Constraints in SQL and Why Are They Used?**

 **Ans-**

 In SQL, constraints are rules applied to table columns that control what kind of data can be stored. They are used to maintain data integrity, accuracy, and reliability in a database.

 **Key Types of Constraints**


*   NOT NULL → Ensures a column cannot have empty values.
*   UNIQUE → Guarantees all values in a column are different.
*   PRIMARY KEY → Uniquely identifies each row in a table.
*   FOREIGN KEY → Links one table to another by referencing a primary key
*   CHECK → Ensures a condition is true (e.g., age > 18).
*   DEFAULT → Provides a default value if none is given.

**Why They Are Used**


*   Prevent invalid or duplicate data.
*   Enforce relationships between tables.
*   Ensure business rules are followed.
*   Improve consistency and reliability of stored information.
